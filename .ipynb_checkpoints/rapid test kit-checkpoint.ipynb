```python
## Imports
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage.filters import threshold_otsu, threshold_local
from skimage import data
from skimage import data, transform, exposure
from skimage.exposure import histogram
from skimage.color import label2rgb
import matplotlib.patches as mpatches
from skimage.measure import label, regionprops
import scipy.ndimage as ndimage
import scipy.ndimage.filters as filters
from skimage.morphology import closing, erosion, dilation
from skimage.io import imread, imsave, imshow, show, imread_collection, imshow_collection
from skimage.exposure import histogram
import skimage.filters
from skimage import feature
from skimage.util import compare_images
from scipy import ndimage as ndi
```

```python
## Viewing the image
image_path = "./images/rapid_test_kit_5.jpeg";
image = imread(image_path)
plt.figure(figsize=(20,10))
plt.imshow(image)
plt.axis('off')
plt.title("Hiv test kit - Negative")
plt.tight_layout();
plt.show();
```

```python
## Viewing the different channels
image_red = image[:,:,0]
image_green = image[:,:,1]
image_blue = image[:,:,2]
fig,axes = plt.subplots(1,3,figsize=(10,5))
axes[0].imshow(image_red);
axes[0].set_title("red");
axes[1].imshow(image_green);
axes[1].set_title("green");
axes[2].imshow(image_blue);
axes[2].set_title("blue");
for a in axes:
    a.axis('off')
plt.tight_layout();
plt.show();
```

```python
gray = skimage.color.rgb2gray(image_green)
dilated_median_blurred = skimage.filters.median(dilation(dilation(dilation(gray))))

#image_blue = imread(image_path)[:,:,1]
#original = skimage.color.rgb2gray(image_blue)

diff = (255 - compare_images(dilated_median_blurred, gray, method='diff'));

hist, hist_centers = histogram(gray)

#local_image = skimage.color.rgb2gray(imread("./images/rapid_test_kit_4.jpeg"))
#local_thresh = threshold_local(local_image, 99, offset=100)
#binary_local = skimage.filters.median(dilation(local_image))

#local_image[local_image >= local_thresh] = 100
#thresh = 75
 
#diff = skimage.feature.canny(diff)
#diff = erosion(diff)
#diff = erosion(diff)
#diff = erosion(diff)
thresh = threshold_otsu(gray)
#local_thresh = threshold_local(diff, 25, offset=10)
print(thresh)
#thresh = 254.9
#diff = diff > thresh
#thresh = 240
#diff = diff - local_thresh
#diff = diff < thresh
gray[gray < thresh] = 1
gray[gray >= thresh] = 0
fig,ax = plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(gray,cmap='gray')
ax[1].plot(hist_centers,hist,lw=2)
plt.tight_layout()
plt.show()




#plt.figure(figsize=(20,10))
#plt.imshow(gray,cmap='gray')
#plt.show()


```

```python
## perform a closing step.
img = closing(gray)

## perform 3 consecutive erosions, this helps to delineate the packets which are touching 
## the flooring lines.
e1 = erosion(img)
#e2 = dilation(e1)
#e3 = dilation(e2)

fig,axes = plt.subplots(1,3,figsize=(20,10))
axes[0].imshow(e1,cmap="gray");
axes[0].set_title("first erosion - lots of lines");
axes[1].imshow(e2,cmap="gray");
axes[1].set_title("second erosion - some lines disappearing");
axes[2].imshow(e3,cmap="gray");
axes[2].set_title("third erosion - all lines gone");
for a in axes:
    a.axis('off')
plt.tight_layout();
plt.show();


## call label on the resulting image.
labelled = label(e1)

## show the labels with rectangles, overlay on the original image, filtering rectangles that
## are greater than 1000 pixels, this removes one or two noisy features.
image_label_overlay = label2rgb(labelled, image=e1, bg_label=0)
fig, ax = plt.subplots(figsize=(20,10))
ax.imshow(gray)

for region in regionprops(labelled):
    if region.area >= 10:
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                          fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)

ax.set_axis_off()
plt.title("segmented")
plt.tight_layout()
plt.show()
```
